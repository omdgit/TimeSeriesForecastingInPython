1. unconditional mean   
2. variance constant  
3. covariance depends on time j that has elapsed between observations, not on reference period (regardless of 2 or 3 periods apart)  
    For example: covariance between t<sub>4</sub> and t<sub>8</sub>  same as for t<sub>9</sub> and t<sub>13</sub> where j=4

### Simulate an AR(1) process and use Auto-ARIMA to find it

```python
# Create AR(1) time series with white noise
np.random.seed(42)

white_noise = np.random.standard_normal(1500)
constant = 1
coefficient = 0.7
ar1 = []

for i in np.arange(0, 1500):
    if i == 0:
        ar1.append(constant + coefficient * 1 + white_noise[i])
    else:
        ar1.append(constant + coefficient * ar1[i-1] + white_noise[i])
```

```python
tt.plot_for_stationary(pd.Series(ar1))
tt.adf(ar1)
```

```python
ps = range(0, 6, 1)
qs = range(0, 6, 1)

order_list = list(product(ps, qs))

arma_auto = tt.optimize_ARMA(ar1, order_list) 
arma_auto.head(10)
```

```python
arma_auto_model = sm.tsa.ARIMA(ar1, order=(1,0,0))
arma_auto_model_fit = arma_auto_model.fit()
display(arma_auto_model_fit.summary())
arma_auto_model_fit.plot_diagnostics(figsize=(10,8))
arma_auto_residuals = arma_auto_model_fit.resid
tt.ljungbox(np.array(arma_auto_residuals), lags=12)
```
